```python
from bs4 import BeautifulSoup
from splinter import Browser
import pymongo
import pandas as pd
import requests
```

```python
url ="https://mars.nasa.gov/news/"
# create soup
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
#print(soup.prettify())
```

```python
#latest_news = soup.find_all("div", class_="slide")
#print (latest_news)
```

```python
#get Latests News
latest_news = soup.find_all("div", class_="slide")
for news in latest_news:
    try:  
        news_title = news.find("div", class_="content_title")
        title_text = news_title.find("a").text
        news_article =  news.find("div", class_="rollover_description")
        article = news_article.find("div", class_="rollover_description_inner").text
        news_link = news.a["href"]
        # if news title exist print
        if (news_title):
            print ("------------------------------")
            print (f"Title: {title_text}")
            print (f"Article: {article}")
            print (f"Link: {news_link}")
    except AttributeError as e:
        print(e)
```

```python
#Splinter
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
url = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
browser.visit(url)
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
#print(soup.prettify())
test = soup.find_all("a", class_="fancybox")
#print(test)
```

```python
# Iterate through More Button Clicks - 1
button_clicks = 3
for x in range(button_clicks):
    
    # HTML object
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    featured_img = soup.find_all("a", class_="fancybox")

    # Iterate through each book
    for img in featured_img:
        # Use Beautiful Soup's find() method to navigate and retrieve attributes
        featured_image_link = img['data-fancybox-href']
        print('-----------')
        print('https://www.jpl.nasa.gov' + featured_image_link)
        
    # Click the 'MORE' button on each page
    try:
        browser.click_link_by_partial_text('MORE')
    except:
        print("Scraping Complete")
```
