```python
from bs4 import BeautifulSoup
from splinter import Browser
from pprint import pprint
import pymongo
import pandas as pd
import requests
```

```python
url ="https://mars.nasa.gov/news/"
# create soup
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
#get Latests News
latest_news = soup.find_all("div", class_="slide")
for news in latest_news:
    try:  
        news_title = news.find("div", class_="content_title")
        title_text = news_title.find("a").text
        news_article =  news.find("div", class_="rollover_description")
        article = news_article.find("div", class_="rollover_description_inner").text
        news_link = news.a["href"]
        # if news title exist print
        if (news_title):
            print ("------------------------------")
            print (f"Title: {title_text}")
            print (f"Article: {article}")
            print (f"Link: {news_link}")
    except AttributeError as e:
        print(e)
```

```python
#Mars Mars Hemispheres
hemisphere_image_urls = []

cerberus_url = "https://astrogeology.usgs.gov/search/map/Mars/Viking/cerberus_enhanced"
schia_url = "https://astrogeology.usgs.gov/search/map/Mars/Viking/schiaparelli_enhanced"
syrtis_url = "https://astrogeology.usgs.gov/search/map/Mars/Viking/syrtis_major_enhanced"
valles_url = "https://astrogeology.usgs.gov/search/map/Mars/Viking/valles_marineris_enhanced"
hem_urls = [cerberus_url,schia_url,syrtis_url,valles_url]

#loop through each url and get needed data
for url in hem_urls:
    # create soup
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    img_container = soup.find_all('div', class_="wide-image-wrapper")
    for hem_img in img_container:
        img = hem_img.find("li")
        full_hem_img = img.a["href"]
        title = soup.find('h2', class_="title").text

        try:
            hemisphere_image_urls.append({"title": title, "img_url": full_hem_img})
        except:
            print("Append Error!")
print ("hemisphere_image_urls")
pprint (hemisphere_image_urls)
```
